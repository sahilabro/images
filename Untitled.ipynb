We will be using python to make wrapper for imagemagick, which is a CLI utility to resize images, or convert them into different types

```python
#this function will take a jpg path and resize the image
#the root path is same folder as python script
#the path must include extension (i.e. for example /rose.jpg)
#Import all the required libraries

import os
import json 
```

Takes optional argument of imagefiletype. If not given it will output a jpg file.

Takes 3 arguments: path, resizeAttribute, outputfilename. And optional argument ^


```python
def resizeMyImage(path,resizeAttribute,outputfilename, **kwargs):

    #check for optional argument (if output filetype other than jpg)
    if 'imagefiletype' in kwargs:
        imagefiletype = kwargs['imagefiletype']
    else:
        imagefiletype = 'jpg'
    
    #for debugging if anything goes wrong
    cliOutput = os.system("magick convert %s -resize %s %s"%(path,resizeAttribute,outputfilename+'.'+imagefiletype))
    
    return 'New image is called %s\nIt is situated in the same folder as this script.'%(outputfilename)

```

This takes a json file and parses it into our resizeMyImage function. Please note the JSON file must have the following structure:

{"images":
        [
            {"path": "rose.jpg", "resizeAttribute": "3%", "outputfilename": "my2dNewFile"},
            {"path": "rose.jpg", "resizeAttribute": "30%", "outputfilename": "my2d2NewFile", "imagefiletype":"png"}
        ]
    }
    
    
the imagefiletype is optional. It will output jpg if not given.


```python
def multiImagefromJson(jsonFilePath):

    #First things first. Open the file and load it into python dictionary:
    with open(jsonFilePath, 'r') as f:
        imagesTobeParsed = json.load(f)

    for image in imagesTobeParsed['images']:
        imagefiletype = 'jpg' if 'imagefiletype' not in image else image['imagefiletype']
        resizeMyImage(image['path'],image['resizeAttribute'],image['outputfilename'],imagefiletype='.'+imagefiletype)
        
    return('resizing and/or converting images taken from %s'%(jsonFilePath))
```

# What happens when we push this onto a production environment

This code does not use any external libraries, therefore no libraries will be needed when thinking of containerisation and packaging the application.

However, we will need to make sure the imagick dependency is met. This application can be packaged as a docker image.

To make sure this application is scalable when dealing with multiple queries at a time, a new instance for each x amount of users.

The variable x here will be dependent on how many users one docker image can handle, this can be tested once a docket image is deployed on AWS. This may be dependent on image density and filesizes. Also dependent on conversion type.



Moreover, for scalability, and if this was to be deployed on the web and we were to introduce scheduled jobs, we can use MongoDB.

MongoDB is the perfect database for this tool as it deals with JSON objects - it is a non relational database (noSQL)

MongoDB can be deployed in a separate docker instance, or we can use a similar AWS product. We would have to make sure that we expose MongoDB services to our original docker image.



As MongoDB is a scalable database, we can store monitoring and logging statistics on there for users too.

We may then develop further applications on top of this DB for monitoring, admin and logging purposes. This data can be analysed from a data science angle for commercial use afterwards too.


Furthermore, we can change the application as such that if it is over-loaded, the jobs can be stored as 'scheduled jobs,' on MongoDB - and done as soon as the current jobs are finished. A priority parameter can be introduced too.
